```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
x = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init='k-means++')
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
kmeanss = KMeans(n_clusters=5,init='k-means++',random_state=0)
kmeanss.fit(x)
```

```python
y = kmeanss.fit_predict(x)
```

```python
plt.scatter(x[y==0,0],x[y==0,1],s=100,c='red',label='cluster1')
plt.scatter(x[y==1,0],x[y==1,1],s=100,c='violet',label='cluster2')
plt.scatter(x[y==2,0],x[y==2,1],s=100,c='turquoise',label='cluster3')
plt.scatter(x[y==3,0],x[y==3,1],s=100,c='yellow',label='cluster4')
plt.scatter(x[y==4,0],x[y==4,1],s=100,c='brown',label='cluster5')
plt.scatter(kmeanss.cluster_centers_[:,0],kmeanss.cluster_centers_[:,1],s=200,c='black',label='Centroid')
plt.legend()
plt.show()
```
